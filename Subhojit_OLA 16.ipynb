**Task 3**

Print the first five rows of the DataFrame named "Actual" using the provided URL.

Actual: https://raw.githubusercontent.com/Invact-Abhay/DOE11/refs/heads/main/SmartChoice%20-%20Actual.csv

```python
Actual=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE11/refs/heads/main/SmartChoice%20-%20Actual.csv")
Actual.head()
```

**Task 4**

Convert all column names in the Actual and Budget DataFrame to lowercase and replace spaces with underscores.


```python
Budget.columns=Budget.columns.str.lower().str.replace(" ","_")
Actual.columns=Actual.columns.str.lower().str.replace(" ","_")
```

```python
Budget.columns
```

```python
Actual.columns
```

**Task 5**

Group the Actual Data Frame by month, region, factory-code, and model, then sum the actual production column and create a new Data Frame with the result stored in a column named actual. Your output should contain following columns month,region, factory_code, model and actual production


```python
actual=Actual.groupby(['month','region','factory_code','model'])['actual_quantity_produced'].sum().reset_index()
actual
```

**Task 6**

Merge the Budget DataFrame with the Grouped_Actual DataFrame, based on the month, region, factory-code and model columns. Your Output should include the following columns month, region, factory_code, model and actual _production.


```python
merged_table=pd.merge(Budget,actual, on=['month','region','factory_code','model'],how='left')
merged_table
```

**Task 7**

Calculate the variance between actual production and budgeted_quantity for each row in the merged_table DataFrame and store the result in a new column called variance. Your Output should include the following columns month, region, factory_code, model, actual _production and variance.


```python
merged_table['variance']=merged_table['actual_quantity_produced']-merged_table['budgeted_quantity']
merged_table
```

**Task 8**

Create a new column month_number in the merged_table DataFrame by mapping the month names to their corresponding numerical values using a dictionary, and then display the first few rows of the updated DataFrame. Your Output should include the following columns month, region, factory_code, model, actual _production, variance and month_number.


```python
month_mapping={'January':1,
               'February':2,
               'March':3,
               'April':4,
               'May':5,
               'June':6,
               'July':7,
               'August':8,
               'September':9,
               'October':10,
               'November':11,
               'December':12 }
merged_table['month_number']=merged_table['month'].map(month_mapping)
merged_table
```

**Task 9**

Group the merged_table by month and month_number, sum the variance for each group and then sort the resulting monthly_merged_table by month_number in ascending order.  Your Output should include the following columns month, month number and variance.




```python
monthly_merged_table=merged_table.groupby(['month','month_number'])['variance'].sum().reset_index().sort_values('month_number',ascending=True)
monthly_merged_table
```

**Task 10**

Create a bar chart using the monthly_merged DataFrame, where the month column is plotted on the x-axis and the variance column is plotted as the height of the bars, also label the axes and rotate the x-axis labels by 90 degrees. Name the Chart as "Monthly Variance".



```python
plt.bar(x=monthly_merged_table['month'],height=monthly_merged_table['variance'])
plt.xlabel("Month")
plt.ylabel("Variance")
plt.xticks(ticks=monthly_merged_table['month'],rotation=90)
plt.title("Monthly Variance")
plt.show()
```

**Task 11**

Create a region_wise_variance DataFrame that groups the merged_table by 'region' and calculates the sum of the 'variance' for each region, then resets the index. Your Output should include the following columns region and variance.





```python
region_wise_variance=merged_table.groupby('region')['variance'].sum().reset_index()
region_wise_variance
```

**Task 12**

Create a bar chart using the region_wise_variance DataFrame, where the 'region' column is plotted on the x-axis and the 'variance' column represents the height of the bars, while also labeling the axes appropriately. Name the chart as "Region-wise Variance".


```python
plt.bar(x=region_wise_variance['region'],height=region_wise_variance['variance'])
plt.xlabel("Region")
plt.ylabel("Variance")
plt.title("Region-wise Variance")
plt.show()
```

**Task 13**

DataFrame called model_wise_variance_table that summarizes the total variance of production quantity by model from the merged_table DataFrame, using the group by function to aggregate the 'variance' column.Your Output should include the following columns model and variance.



```python
model_wise_variance=merged_table.groupby('model')['variance'].sum().reset_index()
model_wise_variance
```

**Task 14**

Create a bar chart to visualize the model-wise variance using the model_wise_variance DataFrame, where the ‚Äúmodel‚Äù names are on the x-axis and the ‚Äúvariance‚Äù values are represented as the height of the bars.

Additionally, label the axes and rotate the x-axis labels by 90 degrees for better readability. Name the chart as " Model-wise Variance".



```python
plt.bar(x=model_wise_variance['model'],height=model_wise_variance['variance'])
plt.xlabel("Model")
plt.ylabel('Variance')
plt.xticks(ticks=model_wise_variance['model'],rotation=90)
plt.title("Model-wise Variance")
plt.show()
```
